```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('AB_NYC_2019.csv')
df
```

```python
df['price'].replace(0,np.nan,inplace=True)
df
```

```python
df['price'].isnull().sum()
```

```python
df['price'].mean()
```

```python
df['price'].median()
```

```python
zero_percentile= df['price'].quantile(0)
```

```python
nintynine_percentile = df['price'].quantile(0.99)
```

```python
df_no_outlier = df[(df['price']>zero_percentile) & (df['price']<nintynine_percentile)]
df_no_outlier
```
